```python
from io import TextIOWrapper
import re


class File:
    filename: str
    _f: TextIOWrapper
    _lines: str = ""
    _content: str = ""

    def __init__(self, filename: str):
        self.filename = filename

        try:
            self._f = open(self.filename, "r", encoding="utf-8")
        except:
            raise
    
    def get_lines(self):
        self._f.seek(0)
        if self._lines == "":
            self._lines = self._f.readlines()

        return self._lines

    def get_content(self):
        self._f.seek(0)
        if self._content == "":
            self._content = self._f.read()

        return self._content

    def count(self) -> str:
        lines = len(self.get_lines())
        letters = len(self.get_content())
        words = len(re.split("\s+", self.get_content()))

        return (lines, letters, words)

    def head(self, n: int):
        return self.get_lines()[:n]

    def tail(self, n: int):
        return self.get_lines()[-n:]

    def search(self, str: str):
        lines: list[str] = []

        for line in self.get_lines():
            if str in line:
                lines.append(line)

        return lines

    def replace(self, str1: str, str2: str):
        return self.get_content().replace(str1, str2)

    def close(self):
        self._f.close()
```

```python
f = File("iverson.txt")

print(f.head(3))
print(f.tail(3))
print(f.search("Jey"))
print(f.count())
print(f.replace("Jey", "Owen"))

f.close()
```
